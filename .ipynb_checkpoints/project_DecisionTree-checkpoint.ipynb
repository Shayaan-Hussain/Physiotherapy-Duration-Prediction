```python
#Import the required modules and Linear regression class from model.py
import pandas as pd
import numpy as np
from model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read the dataset
physio = pd.read_excel('physio.xlsx')
physio.head()
```

```python
#Column and Row adjustments
physio.columns = physio.iloc[0]
physio.drop(0, inplace=True)
physio = physio[physio.columns[1:]]
physio.head()
```

```python
#Converting intensity of pain and treatment duration to integer values from string
physio['Intensity of pain'] = physio['Intensity of pain'].apply(lambda x:int(x.split()[0]))
physio['Duration of treatment'] = physio['Duration of treatment'].apply(lambda x:int(x.split()[0]))
physio.head()
```

```python
#Splitting diagnosis into separate columns
'''diagnosis = pd.get_dummies(physio['Diagnosis'], drop_first=True)
physio = pd.concat([physio, diagnosis], axis = 1)
physio.drop(['Diagnosis'], axis = 1, inplace = True)
physio.head()'''

diagnosis = list(physio['Diagnosis'].unique())
physio['Diagnosis'].replace(diagnosis, range(len(diagnosis)), inplace=True)
physio.head()
```

```python
#Assigning integer values to treatment approach and pain duration
physio['Treatment approach'].replace(['manual','mechanical','manual and mechanical'], [0,0.5,1], inplace=True)
physio['Duration of pain'].replace(['acute','subacute','chronic'], [0,0.5,1], inplace=True)
physio.head()
```

```python
#Checking for null values
physio.isnull().sum()
```

```python
#Since age has many null values, drop age
#We dont need affected body part column since the prediction is based on the diagnosis
physio.drop(['Age','Affected body part'], axis=1, inplace=True)
physio.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(max_depth=8)
```

```python
x = physio.drop(['Duration of treatment'], axis=1)
y = physio['Duration of treatment']
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
model.fit(xtrain,ytrain)
```

```python
pred = model.predict(xtest)
```

```python
model.score(xtest,ytest), model.score(xtrain,ytrain)
```

```python
pred = model.predict(x)
pred = pd.DataFrame({'Predictions':pred})
pred.head()
```

```python
pd.concat([x,y,pred], axis=1)
```

```python
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image  
import pydotplus

dot_data = StringIO()
export_graphviz(model, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = list(x.columns),class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('diabetes.png')
Image(graph.create_png())
```
